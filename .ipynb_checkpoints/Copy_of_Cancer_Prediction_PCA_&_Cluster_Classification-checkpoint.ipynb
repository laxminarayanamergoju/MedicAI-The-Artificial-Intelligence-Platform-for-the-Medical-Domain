# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import SpectralClustering, KMeans
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import adjusted_rand_score, accuracy_score, confusion_matrix, f1_score, precision_score, recall_score, roc_curve, auc, ConfusionMatrixDisplay
from scipy.optimize import minimize
from deap import base, creator, tools, algorithms
import warnings

warnings.filterwarnings("ignore")
```

# 1. Downloading and analyzing data

```python
data = pd.read_csv('cancer.csv')
data.head()
```

## Distribution of the target variable

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Diagnosis', data=data, palette='Set2')
plt.title('Distribution of the target variable (Diagnosis)')
plt.xlabel('Diagnosis (0 = Healthy, 1 = Sick)')
plt.ylabel('Count')
plt.show()
```

# 2. Visualization of pairwise dot plots for the distribution of the target variable with features

```python
# Delete the target variable (Diagnosis)
X = data.drop(columns='Diagnosis')

# Target variable (0 - No Cancer, 1 - Cancer)
y = data['Diagnosis']
```

```python
def visualize_pairwise_distributions(df, features_per_block=5):
    for i in range(0, len(df.columns)-1, features_per_block):
        sns.pairplot(
            df,
            hue='Diagnosis',  # Використовуємо 'Diagnosis' як таргет
            vars=df.columns[i:i+features_per_block],
            palette='Set1',
        )
        plt.show()

visualize_pairwise_distributions(data)
```

# 3. Clustering by Spectral Clustering, KMeans and Gaussian mixture model

```python
clustering_methods = {
    "Spectral Clustering": SpectralClustering(n_clusters=2, affinity='nearest_neighbors', random_state=42),
    "KMeans": KMeans(n_clusters=2, random_state=42),
    "Gaussian Mixture": GaussianMixture(n_components=2, random_state=42),
}

cluster_results = {}
for name, model in clustering_methods.items():
    labels = model.fit_predict(X)
    score = adjusted_rand_score(y, labels)
    cluster_results[name] = (labels, score)
    print(f"{name} Adjusted Rand Score: {score:.2f}")


# Visualization of clustering results
for name, (labels, _) in cluster_results.items():
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=labels, palette="viridis", alpha=0.7)
    plt.title(f"{name} Clustering")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

# 4-5. Data dimensionality reduction using PCA and visualization of class distribution in a new feature space

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette="Set1", alpha=0.7)
plt.title("PCA: Data Distribution in Reduced Space")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
# Percentage of variance explained
plt.figure(figsize=(8, 5))
plt.bar(range(1, pca.n_components_ + 1), pca.explained_variance_ratio_)
plt.title("Explained Variance by Principal Components")
plt.xlabel("Principal Component")
plt.ylabel("Variance Ratio")
plt.xticks([1, 2], ["1", "2"])
plt.show()
```

# 6. Classification by the logistic regression method

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
log_reg = LogisticRegression(solver='lbfgs', max_iter=1000)
log_reg.fit(X_train, y_train)

y_pred = log_reg.predict(X_test)
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, y_pred):.2f}")

plt.figure(figsize=(8, 6))
plt.scatter(range(len(y_test)), y_test, color='blue', label='Actual', alpha=0.6)
plt.scatter(range(len(y_pred)), y_pred, color='red', label='Predicted', alpha=0.6)
plt.title("Logistic Regression: Actual vs Predicted")
plt.xlabel("Sample Index")
plt.ylabel("Class Label")
plt.legend()
plt.show()
```

# 7. Logistic regression with gradient descent optimization

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def cost_function(theta, X, y):
    m = len(y)
    h = sigmoid(np.dot(X, theta))
    epsilon = 1e-5
    return -(1/m) * np.sum(y * np.log(h + epsilon) + (1 - y) * np.log(1 - h + epsilon))

def gradient(theta, X, y):
    m = len(y)
    h = sigmoid(np.dot(X, theta))
    return (1/m) * np.dot(X.T, h - y)

initial_theta = np.zeros(X_train.shape[1])
result = minimize(fun=cost_function, x0=initial_theta, args=(X_train, y_train), method='BFGS', jac=gradient)
optimal_theta = result.x

y_pred_gd = np.round(sigmoid(np.dot(X_test, optimal_theta)))

# Logistic regression from the sklearn library
log_reg = LogisticRegression(solver='lbfgs', max_iter=1000)
log_reg.fit(X_train, y_train)

y_pred_sklearn = log_reg.predict(X_test)

# Model accuracy
accuracy_gd = accuracy_score(y_test, y_pred_gd)
accuracy_sklearn = accuracy_score(y_test, y_pred_sklearn)

print(f"Gradient Descent Logistic Regression Accuracy: {accuracy_gd:.2f}")
print(f"Sklearn Logistic Regression Accuracy: {accuracy_sklearn:.2f}")

# Accuracy comparison
plt.bar(['Gradient Descent', 'Sklearn'], [accuracy_gd, accuracy_sklearn], color=['blue', 'green'])
plt.title('Accuracy Comparison')
plt.ylabel('Accuracy')
plt.show()
```

# 8. Logistic regression with optimization using a genetic algorithm

```python
def evaluate(individual):
    theta = np.array(individual)
    y_pred_ga = np.round(sigmoid(np.dot(X_train, theta)))
    return (accuracy_score(y_train, y_pred_ga),)

creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)
toolbox = base.Toolbox()

toolbox.register("attr_float", np.random.uniform, -1, 1)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=X_train.shape[1])
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evaluate)

population = toolbox.population(n=100)
result, _ = algorithms.eaSimple(population, toolbox, cxpb=0.5, mutpb=0.2, ngen=40, verbose=False)
best_ind = tools.selBest(population, k=1)[0]

optimal_theta_ga = np.array(best_ind)
y_pred_ga = np.round(sigmoid(np.dot(X_test, optimal_theta_ga)))
print(f"Genetic Algorithm Logistic Regression Accuracy: {accuracy_score(y_test, y_pred_ga):.2f}")
```

# 9. Assessing the quality of models with Confusion Matrix

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, (name, preds) in enumerate([("Gradient Descent", y_pred_gd), ("Logistic Regression", y_pred), ("Genetic Algorithm", y_pred_ga)]):
    cm = confusion_matrix(y_test, preds)
    f1 = f1_score(y_test, preds)
    precision = precision_score(y_test, preds)
    recall = recall_score(y_test, preds)

    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Healthy", "Sick"])
    disp.plot(ax=axes[i], cmap=plt.cm.Blues)
    axes[i].set_title(f"Confusion Matrix: {name}")

    print(f"{name}:")
    print("Confusion Matrix:")
    print(cm)
    print(f"F1 Score: {f1:.2f}, Precision: {precision:.2f}, Recall: {recall:.2f}\n")

plt.tight_layout()
plt.show()
```

## ROC Curve

```python
fpr, tpr, _ = roc_curve(y_test, y_pred_gd)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f"Gradient Descent (AUC = {roc_auc:.2f})")

fpr, tpr, _ = roc_curve(y_test, y_pred_sklearn)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f"Sklearn Logistic Regression (AUC = {roc_auc:.2f})")

fpr, tpr, _ = roc_curve(y_test, y_pred_ga)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f"Genetic Algorithm (AUC = {roc_auc:.2f})")

plt.title("ROC Curve for All Models")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report


df =pd.read_csv('cancer.csv')

# Features and target
X = df.drop(columns=['Diagnosis'])
y = df['Diagnosis']

# Split data into train and test sets (e.g., 60% train, 40% test because data is small)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Initialize Gradient Boosting classifier
model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Train the model on training data
model.fit(X_train, y_train)

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy on test set:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import gradio as gr
import pandas as pd

def predict_cancer(Age, Gender, BMI, Smoking, GeneticRisk, PhysicalActivity, AlcoholIntake, CancerHistory):
    input_data = pd.DataFrame({
        'Age': [Age],
        'Gender': [Gender],
        'BMI': [BMI],
        'Smoking': [Smoking],
        'GeneticRisk': [GeneticRisk],
        'PhysicalActivity': [PhysicalActivity],
        'AlcoholIntake': [AlcoholIntake],
        'CancerHistory': [CancerHistory]
    })
    pred = model.predict(input_data)[0]
    prob = model.predict_proba(input_data)[0][1]
    return f"Prediction: {'Cancer' if pred==1 else 'No Cancer'} (Probability: {prob:.2f})"

interface = gr.Interface(
    fn=predict_cancer,
    inputs=[
        gr.Slider(20, 80, step=1, label="Age"),
        gr.Radio(choices=[("Male", 0), ("Female", 1)], label="Gender"),
        gr.Slider(15, 40, step=0.1, label="BMI"),
        gr.Radio(choices=[("No", 0), ("Yes", 1)], label="Smoking"),
        gr.Radio(choices=[("Low", 0), ("Medium", 1), ("High", 2)], label="Genetic Risk"),
        gr.Slider(0, 10, step=0.1, label="Physical Activity (hours/week)"),
        gr.Slider(0, 5, step=0.1, label="Alcohol Intake (units/week)"),
        gr.Radio(choices=[("No", 0), ("Yes", 1)], label="Personal Cancer History")
    ],
    outputs="text",
    title="Cancer Diagnosis Predictor",
    description="Input patient data to predict cancer diagnosis."
)

interface.launch()
```
